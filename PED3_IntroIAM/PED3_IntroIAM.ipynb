## Arquitecturas de Modelos de Evaluación Integrada (IAM)

## 0. Introducción conceptual

En el bloque teórico hemos estudiado qué son los Modelos de Evaluación Integrada (IAM) y cuál es su papel en la construcción de escenarios climáticos internacionales. En este bloque vamos a dar un paso adicional: comprender cómo funcionan internamente y qué implicaciones tiene su arquitectura.

El objetivo no es aprender economía formal ni derivar ecuaciones complejas. El objetivo es entender las **reglas básicas** que estructuran estos modelos y cómo dichas reglas condicionan los resultados.

---

## 0.1 Tres planos conceptuales

Para evitar confusiones, es importante distinguir tres niveles distintos:

### 1. Arquitectura del modelo

La arquitectura define las **relaciones estructurales** del sistema.  
Es, por así decirlo, el “motor” del modelo.
Esta arquitectura está formada por:

- Las ecuaciones que describen la dinámica climática.
- Las reglas que describen la producción y las emisiones.
- La forma en que los daños afectan a la economía.
- La regla de decisión (cómo se decide mitigar).


---

### 2. Escenario

Es el marco general dentro del cual se toman decisiones.
Un escenario describe un contexto socioeconómico coherente:

- Crecimiento demográfico.
- Evolución económica.
- Desarrollo tecnológico.
- Contexto institucional.

El escenario no es una política concreta.  

El análisis detallado de escenarios se abordará en el siguiente bloque.

---

### 3. Política climática

La política es una intervención deliberada dentro de un escenario, por ejemplo:

- Impuesto al carbono.
- Límite de emisiones.
- Subvenciones.
- Regulaciones.

Una misma política puede producir resultados distintos según el escenario en el que se aplique.

---

## 0.2 Dos grandes arquitecturas de IAM

En este bloque compararemos dos enfoques principales:

### A) Modelos de optimización global

En estos modelos se elige la trayectoria completa de mitigación que minimiza el coste total acumulado a lo largo del tiempo (formalmente, maximiza una función de utilidad social intertemporal).

Su lógica es:

> Si pudiéramos coordinar todas las decisiones presentes y futuras de forma coherente, **¿cuál sería la trayectoria eficiente?**

Estos modelos son normativos: proporcionan una referencia de eficiencia intertemporal.

---

### B) Modelos dinámico–recursivos

En estos modelos las decisiones se toman paso a paso, reaccionando a la situación actual (temperatura, daños, señales económicas).

Su lógica es:

> Dadas ciertas reglas institucionales y tecnológicas, **¿cómo evoluciona el sistema en el tiempo?**

Estos modelos son descriptivos: representan dinámicas bajo reglas dadas, con inercias y restricciones.

---

## 0.4 En qué NO nos vamos a centrar

- No vamos a estudiar teoría económica avanzada.

Nos centraremos en comprender la lógica estructural y en interpretar los resultados.

# 1. Arquitectura del modelo de optimización global

En este modelo suponemos que la trayectoria completa de mitigación se elige considerando simultáneamente el presente y el futuro. 

Es decir, no se decide año a año reaccionando a la situación actual, sino que se calcula de una vez cuál es la senda que minimiza el coste total acumulado.

El modelo tiene cuatro bloques principales:

1. Producción y emisiones  
2. Dinámica climática  
3. Daños climáticos  
4. Coste total acumulado (criterio de decisión)

---

## 1.1 Producción y emisiones

Suponemos que la producción genera emisiones proporcionales:

E_t = sigma * (1 - mu_t) * Y_t

donde:

- Y_t es la producción.
- sigma es la intensidad de carbono.
- mu_t es el nivel de mitigación (entre 0 y 1).
- E_t son las emisiones.

Interpretación:

- Si mu_t = 0 → no hay mitigación.
- Si mu_t = 1 → emisiones eliminadas.
- A mayor mitigación, menores emisiones.

Aquí aparece una regla económica básica:

> Reducir emisiones requiere esfuerzo (mitigación) que afecta al sistema productivo.

---

## 1.2 Dinámica climática simplificada

El carbono se acumula en el sistema:

C_{t+1} = (1 - delta) * C_t + E_t

La temperatura responde al carbono acumulado:

T_{t+1} = T_t + phi * C_t

donde:

- C_t es el stock acumulado de carbono.
- delta es la fracción que se elimina naturalmente.
- T_t es la temperatura media.
- phi es la sensibilidad climática simplificada.

Interpretación:

- Las emisiones actuales afectan al futuro.
- Existe inercia en el sistema climático.

---

## 1.3 Daños climáticos

La temperatura reduce la producción:

Y_neta_t = Y_t * (1 - theta * T_t**2)

donde:

- theta mide la intensidad de los daños.

Interpretación:

- A mayor temperatura, mayor pérdida de producción.
- La forma cuadrática es una simplificación estructural.
- El valor de theta es un supuesto crucial del modelo.

---

## 1.4 Coste de mitigación

Mitigar no es gratuito:

Coste_mit_t = alpha * mu_t**2 * Y_t

donde:

- alpha representa el coste relativo de mitigar.
- El término mu_t**2 indica que mitigar mucho es progresivamente más costoso.

---

## 1.5 Regla central: minimización del coste total acumulado

El modelo elige la trayectoria completa de mu_t que minimiza:

Coste_total = sum( beta**t * (Coste_mit_t + Danos_t) )

donde:

- beta es el factor de descuento.
- Danos_t son las pérdidas asociadas a la temperatura.
- La suma incluye costes presentes y futuros.

Interpretación intuitiva:

El modelo equilibra:

- el coste de mitigar hoy,
- frente a los daños que ocurrirán en el futuro.

Formalmente, esto se denomina maximización de una función de utilidad social intertemporal.

---

## 1.6 Qué implica esta arquitectura

Este modelo supone:

- Coordinación perfecta a lo largo del tiempo.
- Conocimiento completo del futuro.
- Capacidad de comprometer decisiones intertemporales.

Por eso responde a la pregunta:

> ¿Cuál sería la trayectoria eficiente si pudiéramos planificar todo el horizonte temporal de forma coherente?
